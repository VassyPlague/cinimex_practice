```python
import codecs
import pandas as pd

def parse(path):
    
    out = ''
    fileObj = codecs.open(path, "r", "utf_8_sig" ) # открытие с правильной кодировкой
    text = fileObj.readlines() 
    fileObj.close()
    
    for i in text: # удаляем мусорные строки
        if i[0] != '#' and i != '\n':
            out += i
            
    with open('data/parsed/cleaned.txt', 'w', encoding= 'utf-8') as f:
        f.write(out) # сохрянем очищенный файл
        
    data = pd.read_csv('data/parsed/cleaned.txt', sep='\t', header=None).rename(columns = {2 : 'word', 3 : 'label'})
    make_df(data) # функция для csv
    

def make_df(data): 3
    data = data[['word', 'label']]
    i = 0
    texts = []
    labels = []

    text = ''
    label_state = data['label'].iloc[0]

    st_pos = 0
    end_pos = 0

    st_l = []
    end_l = []
    while i < len(data):
        
        if label_state != data['label'].iloc[i]:
            end_pos = i - 1
            st_l.append(st_pos)
            st_pos = i
            end_l.append(end_pos)
            labels.append(label_state)
            label_state = data['label'].iloc[i]
            texts.append(text)
            text = data['word'].iloc[i] + ' '
            
        else:
            text += data['word'].iloc[i] + ' '
            
        i += 1
        
    parsed = pd.DataFrame({'st_pos' : st_l, 'end_pos' : end_l, 'text' : texts, 'label' : labels})
    parsed.to_csv('data/parsed/parsed.csv', index= False, sep = ';')
   
    
```
