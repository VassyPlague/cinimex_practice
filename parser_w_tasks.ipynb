```python
import codecs
import pandas as pd
out = ''
fileObj = codecs.open('data/raw/2017_2.tsv', "r", "utf_8_sig" ) # открытие с правильной кодировкой
text = fileObj.readlines() 
fileObj.close()
```

```python
# основная функция
def parse_w_tasks(lines):
    out = ''
    c_task = '' #current task
    for i in lines:
        words = i.split()
        for j in words:
            if '#Text=СИТУАЦИОННАЯ' in words and 'ЗАДАЧА' in words:
                if j.isdigit():
                    c_task = j
        if i.startswith('#') or i.startswith('\n'):
            continue
        else: out += i[:-2]+'\t'+c_task+'\n'
            #out.append(i)
    return out
```

```python
parse_w_tasks(text)
```

```python
with open('data/parsed/task_parsed.txt', 'w', encoding= 'utf-8') as f:
        f.write(parse_w_tasks(text)) # сохрянем очищенный файл
data = pd.read_csv('data/parsed/task_parsed.txt', sep='\t', header=None)
data = data [[2,3,7]]
data
```
